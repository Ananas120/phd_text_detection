```python
import numpy as np
import argparse
import torch
from src.transform import SSDTransformer
import cv2
from PIL import Image
import matplotlib.pyplot as plt

from src.utils import generate_dboxes, Encoder, colors, coco_classes
from src.model import SSD, ResNet, Textboxes

def predict(input_image, model_name, checkpoint_path, backbone = ResNet(), nms_threshold = 0.5, split = False):
    predictions = []
    
    dboxes = generate_dboxes(model=model_name)
    encoder = Encoder(dboxes)
    
    if model_name == "tb300" or model_name == "tb300_old":
        model = Textboxes(backbone=backbone, num_classes=2, figsize=300)
        figsize = (300,300)
    elif model_name == "tb512" or model_name == "tb512_old":
        model = Textboxes(backbone=backbone, num_classes=2, figsize=512)
        figsize = (512,512)

    checkpoint = torch.load(checkpoint_path, map_location=torch.device('cpu'))
    model_state_dict = {k.replace('module.', ''): v for k, v in checkpoint["model_state_dict"].items()}
    model.load_state_dict(model_state_dict)
    model.eval()

    
    transformer = SSDTransformer(dboxes, figsize, val=True)
    img = Image.open(input_image).convert("RGB")
    width, height = img.size

    if split:
        quarter_width = width // 2
        quarter_height = height // 2

        # Split the image into four quarters
        top_left = img.crop((0, 0, quarter_width, quarter_height))
        top_right = img.crop((quarter_width, 0, width, quarter_height))
        bottom_left = img.crop((0, quarter_height, quarter_width, height))
        bottom_right = img.crop((quarter_width, quarter_height, width, height))

        quarters = [top_left, top_right, bottom_left, bottom_right]
        
        for q, quarter in enumerate(quarters):
            quarter, _, _, _ = transformer(quarter, None, torch.zeros(1,4), torch.zeros(1))
            if torch.cuda.is_available():
                quarter = quarter.cuda()
            with torch.no_grad():
                # Get predictions
                ploc, plabel = model(quarter.unsqueeze(0))
                ploc, plabel = ploc.float(), plabel.float()

                result = encoder.decode_batch(ploc, plabel,nms_threshold , 200)[0]

                loc, label, prob = [r.cpu().numpy() for r in result]

                for loc_, label_, prob_ in zip(loc, label, prob):
                    xmin,ymin, w, h = loc_[0]/2,loc_[1]/2,(loc_[2] - loc_[0])/2,(loc_[3] - loc_[1])/2
                    if q == 1 or q ==3:
                        xmin += 0.5
                    if q == 2 or q ==3:
                        ymin += 0.5
                    xmin, w = xmin*width, w*width
                    ymin, h = ymin*height, h*height

                    predictions.append((xmin, ymin, w, h, prob_))
    else: 
        img, _, _, _ = transformer(img, None, torch.zeros(1,4), torch.zeros(1))
        if torch.cuda.is_available():
            img = img.cuda()
        with torch.no_grad():
            # Get predictions
            ploc, plabel = model(img.unsqueeze(0))
            ploc, plabel = ploc.float(), plabel.float()

            result = encoder.decode_batch(ploc, plabel,nms_threshold , 200)[0]

            loc, label, prob = [r.cpu().numpy() for r in result]

            for loc_, label_, prob_ in zip(loc, label, prob):
                xmin,ymin, w, h = loc_[0] * width, loc_[1] * height, (loc_[2] - loc_[0]) * width, (loc_[3] - loc_[1]) * height
                predictions.append((xmin, ymin, w, h,prob_))

    return predictions

def plot_predictions(input_image, predictions, cls_threshold = 0.3, show_label = True):
    imagecv = cv2.imread(input_image, cv2.IMREAD_COLOR)
    rgb_image = cv2.cvtColor(imagecv, cv2.COLOR_BGR2RGB)
    print("Image width : {} , Image height : {}".format(imagecv.shape[1],imagecv.shape[0]))

    plt.figure(figsize=(12,12))
    colors = plt.cm.hsv(np.linspace(0, 1, 21)).tolist()
    plt.imshow(rgb_image)  # plot the image for matplotlib
    currentAxis = plt.gca()

    for box in predictions:
        xmin, ymin, box_width, box_height, score = box
        if score > cls_threshold:
            label_name = "Text"
            display_txt = '%s: %.2f'%(label_name, score)
            coords = (xmin,ymin), box_width, box_height
            color = colors[1]
            currentAxis.add_patch(plt.Rectangle(*coords, fill=False, edgecolor=color, linewidth=1.5))
            if show_label:
                currentAxis.text(xmin, ymin, display_txt, bbox={'facecolor':color, 'alpha':0.5})

    plt.axis('off')
    currentAxis.set_title('Boxes with scores > {:.2f}'.format(cls_threshold))
    
def flip_and_merge_image(image_path, output_path):
    # Open the image using Pillow
    image = Image.open(image_path)

    # Get the width and height of the image
    width, height = image.size

    # Calculate the center point of the image
    center = width // 2

    # Split the image into left and right halves
    left = image.crop((0, 0, center, height))
    right = image.crop((center, 0, width, height))

    # Flip each half horizontally
#     left = left.transpose(method=Image.FLIP_LEFT_RIGHT)
#     right = right.transpose(method=Image.FLIP_LEFT_RIGHT)

    # Create a new image with the same size as the original
    new_image = Image.new(mode='RGB', size=(width, height))

    # Paste the left and right halves back into the new image
    new_image.paste(right, (0, 0))
    new_image.paste(left, (center, 0))

    # Save the new image to the output path
    new_image.save(output_path)
```

```python
from PIL import Image
import os
import torch
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import random

colors = [
    "brown",
    "tomato",
    "peachpuff",
    "orange",
    "khaki",
    "lawngreen",
    "turquoise",
    "cyan",
    "crimson",
    "blue",
    "violet",
    "hotpink",
    "yellow",
    "yellow"
]


def filename_to_id(coco):
    mapping = {}
    ids = list(coco.imgs.keys())
    for _id in ids:
        path = coco.loadImgs(_id)[0]["file_name"]
        mapping[path] = _id
    return mapping
    
def plot_debug(image_path, figsize,coco = None, normalizer = None,mapping = None, output=None, cell_idx = None, dboxes = None, root = "/mnt/hgfs/shared/coco/", show_gt = False, TBlayer = None):
    
    image = Image.open(os.path.join(root, image_path)).convert("RGB")


    if normalizer is not None:
        image = transforms.Compose([
                    transforms.Resize(figsize),
                    transforms.ToTensor(),
                    normalizer
                ])(image)
    else:
        image = transforms.Compose([
                    transforms.Resize(figsize),
                    transforms.ToTensor(),
                ])(image)
        
    image_pil = transforms.ToPILImage()(image)
    image_pil.save(output)

    fig, ax = plt.subplots()
    ax.imshow(image.permute(1, 2, 0))

    if show_gt:
        imgID = mapping[image_path]
        target = coco.loadAnns(coco.getAnnIds(imgID))
        
        width, height = image.size
        boxes = []

        for annotation in target:
            bbox = annotation.get("bbox")
            boxes.append([bbox[0] / width, bbox[1] / height, (bbox[0] + bbox[2]) / width, (bbox[1] + bbox[3]) / height])

        boxes = torch.tensor(boxes)
        for box in boxes:
            xmin, ymin, xmax, ymax = box*figsize[0]
            box_width = xmax - xmin
            box_height = ymax - ymin
            coords = (xmin,ymin), box_width, box_height

            ax.add_patch(plt.Rectangle(*coords, fill=False, edgecolor='green', linewidth=1.5))
            
    if TBlayer is not None and dboxes is not None:
        cnt = 0
        n_cells = dboxes.feat_size[TBlayer]*dboxes.feat_size[TBlayer]
        if cell_idx is None:
            cell_idx = random.sample(range(n_cells), 1)[0]

        for center in dboxes.debug_centers[TBlayer]:
            ax.plot(center[0]*figsize[0], center[1]*figsize[0], '.', color='red', markersize=6)
            
        for i, box in enumerate(dboxes.debug_boxes[TBlayer]):
            idx = dboxes.debug_idx[TBlayer][i]
            if idx == cell_idx:
                cx, cy, w, h = box[:4]*figsize[0]
                bottom_left_x = cx - w/2
                bottom_left_y = cy - h/2
                ax.add_patch(plt.Rectangle((bottom_left_x,bottom_left_y), w, h, fill=False,
                                           edgecolor = mcolors.CSS4_COLORS[colors[cnt]], linewidth=1.5))
                cnt += 1

    plt.axis("off")
#     if output:
#         plt.savefig(output)
    plt.show()
```

```python
import csv

def read_predictions(model_names, dataset="new_custom_val_v2"):
    input_dir = os.path.join("predictions",dataset)
    predictions_models = [[] for _ in model_names]
    for i, name in enumerate(model_names):
        # Define the name of the CSV file you want to read
        if dataset == "new_custom_val_v2":
            filename = "val_v2_big_" + name + ".csv"
        else:
            filename = name + ".csv"
        filename = os.path.join(input_dir, filename)
        # Open the CSV file in "read" mode
        with open(filename, mode="r", newline="") as csv_file:
            # Create a reader object using the csv library
            reader = csv.DictReader(csv_file)

            # Define an empty list to store the contents of the CSV file
            contents = []

            # Loop through each row in the CSV file
            for row in reader:
                row["image_id"] = int(row["image_id"])
                row["category_id"] = int(row["category_id"])
                row["bbox"] = eval(row["bbox"])
                row["score"] = float(row["score"])

                # Add the row to the contents list
                if len(row["bbox"]) == 4:
                    contents.append(row)
            predictions_models[i] = contents
    return predictions_models
    
```

```python
root = "/mnt/hgfs/shared/coco/new_custom_val_v2/"
annFile = "/mnt/hgfs/shared/coco/annotations/new_custom_val_v2_big.json"
coco = COCO(annFile)
model_names = ["Textboxes"]
preds = read_predictions(model_names)[0]

image_pred = {}
for pred in preds:
    if pred["image_id"] not in image_pred:
        image_pred[pred["image_id"]] = []
    if pred["score"] > 0.2:
        image_pred[pred["image_id"]].append(pred["bbox"])
```

```python
import numpy as np
import argparse
import torch
from src.transform import SSDTransformer
import cv2
from PIL import Image
import matplotlib.pyplot as plt

from src.utils import generate_dboxes, Encoder, colors, coco_classes
from src.model import SSD, ResNet, Textboxes

checkpoint_path_ssd = "run/new_custom_v2/experiment_SSD512_R152_/checkpoint.pth"
checkpoint_path_tb = "run/new_custom_v2/experiment_realTB512_noOffset_trunc_R152_newBig/checkpoint.pth"
checkpoint_paths = [checkpoint_path_ssd, checkpoint_path_tb]

nms_threshold = 0.5
cls_threshold = 0.2

model_ssd = SSD("ssd512", backbone=ResNet("resnet152"), num_classes=2, figsize=512)
model_tb = Textboxes("tb512_trunc", backbone=ResNet("resnet152"), num_classes=2)
models = [model_ssd, model_tb]

for i, checkpoint_path in enumerate(checkpoint_paths):
    checkpoint = torch.load(checkpoint_path, map_location=torch.device('cpu'))
    model_state_dict = {k.replace('module.', ''): v for k, v in checkpoint["model_state_dict"].items()}
    models[i].load_state_dict(model_state_dict)
    models[i].eval()

dboxes = [generate_dboxes(model="ssd512"), generate_dboxes(model="tb512_trunc")]
transformers = [SSDTransformer(dboxes[0], (512, 512), val=True), SSDTransformer(dboxes[1], (512, 512), val=True)]
encoders = [Encoder(dboxes[0]),Encoder(dboxes[1])]

colors = [(0, 255, 0), (255, 0, 0)]

input_image = '/mnt/hgfs/shared/coco/visu/e5c6201d-f1ea30e0-7f836a41-86c7c4e0-9896d337.jpg'
image_name = input_image.split('/')[-1].split(".jpg")[0]
#input_image = "../../original_data/testing/0bad319f-96900a2d-291c4342-28a8d461-36114997.jpg"
#input_image= "images/mytest.jpg"



img = Image.open(input_image).convert("RGB")





for i, model in enumerate(models):
    output_img = cv2.imread(input_image)
    img_trans, _, _, _ = transformers[i](img, img.size, torch.zeros(1,4), torch.zeros(1))
    with torch.no_grad():
        ploc, plabel = model(img_trans.unsqueeze(dim=0))
        result = encoders[i].decode_batch(ploc, plabel, nms_threshold, 20)[0]
        loc, _, prob = [r.cpu().numpy() for r in result]

        loc = loc[np.argwhere(prob > cls_threshold).squeeze(axis=1)]

        if len(loc) > 0:
            height, width, _ = output_img.shape
            loc[:, 0::2] *= width
            loc[:, 1::2] *= height
            loc = loc.astype(np.int32)
            for box in loc:
                xmin, ymin, xmax, ymax = box
                cv2.rectangle(output_img, (xmin, ymin), (xmax, ymax), colors[i],2)

    if i == 0:
        output = "images/{}_output_ssd.jpg".format(image_name)
    else:
        output = "images/{}_output_tb.jpg".format(image_name)

    cv2.imwrite(output, output_img)
```
