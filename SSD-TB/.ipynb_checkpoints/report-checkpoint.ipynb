```python
experimen_name = "normal"

dataset = "custom"
experiment_dir = os.path.join("run",dataset, "experiment_{}".format(experimen_name))
assert os.path.exists(experiment_dir)

report_dir = os.path.join(experiment_dir, "report")
if not os.path.exists(report_dir):
    os.mkdir(report_dir)
```

# Evaluation Report

```python
model = Textboxes(backbone=ResNet(), num_classes=2)
```

## Train vs Test Loss

```python
with open(os.path.join(experiment_dir, "train_loss.txt"), 'r') as f:
    data_train = f.readlines()
with open(os.path.join(experiment_dir, "test_loss.txt"), 'r') as f:
    data_test = f.readlines()

x1 = []
y1 = []
for line in data_train:
    epoch, value = line.strip().split(' - ')
    x1.append(int(epoch.split()[1]))
    y1.append(float(value))

x2 = []
y2 = []
for line in data_test:
    epoch, value = line.strip().split(' - ')
    x2.append(int(epoch.split()[1]))
    y2.append(float(value))
    
# plot the data from both files on the same graph with a legend
plt.plot(x1, y1, label='Training')
plt.plot(x2, y2, label='Testing')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title("Train vs Test loss")
plt.savefig(report_dir + "/loss.png")
plt.show()
```

## mAP evolution on Test set

```python
with open(os.path.join(experiment_dir, "test_mAP.txt"), 'r') as f:
    data = f.readlines()
    
x = []
y = []
for line in data:
    epoch, value = line.strip().split(' - ')
    x.append(int(epoch.split()[1]))
    y.append(float(value))
plt.plot(x, y)

plt.title("mAp evolution")
plt.xlabel('Epoch')
plt.ylabel('mAP')
plt.savefig(report_dir + "/map.png")
plt.show()
```

```python
test_params = {"batch_size": 4,
               "shuffle": False,
               "drop_last": False,
               "num_workers": 3,
               "collate_fn": collate_fn}

dboxes = generate_dboxes(model="tb")
encoder = Encoder(dboxes)
test_set = CustomDataset("/mnt/hgfs/shared/coco", "val", SSDTransformer(dboxes, (300, 300), val=True))
test_loader = DataLoader(test_set, **test_params)

checkpoint = torch.load(os.path.join(experiment_dir,"checkpoint.pth"), map_location=torch.device('cpu'))
model_state_dict = {k.replace('module.', ''): v for k, v in checkpoint["model_state_dict"].items()}
model.load_state_dict(model_state_dict)

model.eval()
detections = []

cm = ConfusionMatrix(1)

category_ids = test_loader.dataset.coco.getCatIds()

nms_threshold = 0.5

for nbatch, (img, img_id, img_size, gloc, _) in enumerate(test_loader):
    print("Parsing batch: {}/{}".format(nbatch, len(test_loader)), end="\r")
    if torch.cuda.is_available():
        img = img.cuda()
    with torch.no_grad():
        # Get predictions
        ploc, plabel = model(img)
        ploc, plabel = ploc.float(), plabel.float()
        
        assert ploc.shape[0] == gloc.shape[0]
        batch_size = ploc.shape[0]
        

        for idx in range(batch_size):
            ploc_i = ploc[idx, :, :].unsqueeze(0)
            plabel_i = plabel[idx, :, :].unsqueeze(0)
            
            gloc_i = np.array(gloc[idx, :, :])
            gloc_i = gloc_i[np.any(gloc_i != 0, axis=1)]
            
            try:
                result = encoder.decode_batch(ploc_i, plabel_i, nms_threshold, 200)[0]
            except Exception:
                print(traceback.format_exc())
                print("No object detected in idx: {}".format(idx))
                continue
                
            cm_detections = []
            cm_labels = []

            height, width = img_size[idx]
            
            loc, label, prob = [r.cpu().numpy() for r in result]
            for loc_, label_, prob_ in zip(loc, label, prob):
                detections.append([img_id[idx], loc_[0] * width, loc_[1] * height, (loc_[2] - loc_[0]) * width,
                                   (loc_[3] - loc_[1]) * height, prob_,
                                   category_ids[label_ - 1]])
                cm_detections.append([loc_[0] * width, loc_[1] * height, loc_[2] * width, loc_[3] * height, prob_,
                                   0])
                
            for loc in gloc_i:
                cm_labels.append([0,loc[0] * width, loc[1] * height, loc[2] * width, loc[3] * height])
                
            cm_detections = np.array(cm_detections, dtype=np.float32)
            cm_labels = np.array(cm_labels, dtype=np.float32)
            cm.process_batch(cm_detections,cm_labels)


detections = np.array(detections, dtype=np.float32)

coco_eval = COCOeval(test_loader.dataset.coco, test_loader.dataset.coco.loadRes(detections), iouType="bbox")
coco_eval.evaluate()
coco_eval.accumulate()
coco_eval.summarize()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay(confusion_matrix=cm.return_matrix(), display_labels= ["Text", "Background"])
disp.plot()
```

```python
ious = np.arange(0.5,1,0.05)

kept_ious = [0,3,7]

precision = coco_eval.eval["precision"][kept_ious,:,0,0,2]
recall = np.arange(0,1.01,0.01)

fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharex=True)

for i in range(3):
    axs[i].plot(recall, precision[i])
    axs[i].set_title("IOU {:.2f}".format(ious[kept_ious[i]]))
    axs[i].set_xlabel("Recall")
    axs[i].set_ylabel("Precision")

fig.suptitle('PR curve at different thresholds')
```
